```python
import pandas as pd
import pymannkendall as mk

# Create a sample DataFrame (you can replace this with your own data)
data = {'year': [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008],
        'value': [31, 29, 28, 28, 27, 26, 26, 27]}

df = pd.DataFrame(data)

# Set the index of the DataFrame to be the 'year' column
df.set_index('year', inplace=True)

# Perform the Mann-Kendall trend test on the 'value' column
result = mk.original_test(df['value'])

# Print the result
print(result)

# Interpret the result
trend, h, p, z, Tau, s, var_s, slope, intercept = result

if h:
    trend_type = "increasing" if slope > 0 else "decreasing"
    print(f"The trend is {trend_type} with a slope of {slope} and p-value of {p}.")
else:
    print(f"No significant trend detected (p-value: {p}).")
```
